```python
import pandas as pd

file_paths = {
    'control_message_2024': r"C:\Users\admin\Downloads\Dataset #1\ocean_data_challenge_F1_data\races_2024\control_message_2024.csv",
    'lap_2024': r"C:\Users\admin\Downloads\Dataset #1\ocean_data_challenge_F1_data\races_2024\lap_2024.csv",
    'position_2024': r"C:\Users\admin\Downloads\Dataset #1\ocean_data_challenge_F1_data\races_2024\position_2024.csv",
    'result_2024': r"C:\Users\admin\Downloads\Dataset #1\ocean_data_challenge_F1_data\races_2024\result_2024.csv",
    'session_status_2024': r"C:\Users\admin\Downloads\Dataset #1\ocean_data_challenge_F1_data\races_2024\session_status_2024.csv",
    'track_status_2024': r"C:\Users\admin\Downloads\Dataset #1\ocean_data_challenge_F1_data\races_2024\track_status_2024.csv",
    'weather_2024': r"C:\Users\admin\Downloads\Dataset #1\ocean_data_challenge_F1_data\races_2024\weather_2024.csv",
    'car_data_2024': r"C:\Users\admin\Downloads\Dataset #1\ocean_data_challenge_F1_data\races_2024\car_data_2024.csv",
    'circuit_2024': r"C:\Users\admin\Downloads\Dataset #1\ocean_data_challenge_F1_data\races_2024\circuit_2024.csv"
}

output_folder = r"C:\Users\admin\Desktop\work"

def analyze_data(df, name):
    print(f"\n{'='*20} Analysis for {name} {'='*20}")
    
    missing_data = df.isnull().sum()
    missing_data = missing_data[missing_data > 0]
    if missing_data.empty:
        print("\nNo missing data found.")
    else:
        print(f"\nMissing Data for {name}:")
        print(missing_data)

    print(f"\nBasic Statistics for {name}:")
    print(df.describe())

    print(f"\nData Types for {name}:")
    print(df.dtypes)

for name, path in file_paths.items():
    df = pd.read_csv(path)
    analyze_data(df, name)

def fill_missing_data(df):
    numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
    df[numeric_cols] = df[numeric_cols].fillna(0)
    
    time_cols = [col for col in df.columns if 'Time' in col or 'Date' in col or 'LapTime' in col]
    for col in time_cols:
        if df[col].dtype == 'object':
            df[col] = df[col].fillna('0 day 00:00:00')
    
    categorical_cols = df.select_dtypes(include=['object']).columns
    for col in categorical_cols:
        if col not in time_cols:  # We already handled time-like objects
            df[col].fillna('0', inplace=True)
    
    return df

for name, path in file_paths.items():
    df = pd.read_csv(path)
    print(f"\nFilling missing data for {name}...")
    filled_df = fill_missing_data(df)
    
    output_file_path = f"{output_folder}\\{name}_cleaned.csv"
    filled_df.to_csv(output_file_path, index=False)
    print(f"Saved cleaned file for {name} to {output_file_path}")
```
