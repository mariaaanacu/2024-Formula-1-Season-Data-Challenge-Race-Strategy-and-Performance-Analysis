```python
import pandas as pd

file_path = r'C:\Users\admin\Desktop\work\lap_2024_cleaned.csv'
output_path = r'C:\Users\admin\Desktop\work\lap_2024_converted_seconds.csv'

df = pd.read_csv(file_path)

df['PitInTime'] = pd.to_timedelta(df['PitInTime'], errors='coerce')
df['PitOutTime'] = pd.to_timedelta(df['PitOutTime'], errors='coerce')

def timedelta_to_seconds(td):
    if pd.isnull(td):
        return None
    hours = td.components.hours
    minutes = td.components.minutes
    seconds = td.components.seconds + td.components.milliseconds / 1000
    total_seconds = (hours * 3600) + (minutes * 60) + seconds
    return total_seconds

df['PitInTime'] = df['PitInTime'].apply(timedelta_to_seconds)
df['PitOutTime'] = df['PitOutTime'].apply(timedelta_to_seconds)

df['PitTimeSeconds'] = df['PitOutTime'] - df['PitInTime'].shift(1)

df.to_csv(output_path, index=False)

print(f"File successfully exported to: {output_path}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = r'C:\Users\admin\Desktop\work\lap_2024_converted_seconds.csv'
output_folder = r'C:\Users\admin\Desktop\Formula1_output'  

df = pd.read_csv(file_path)

df['PitOutTime'] = pd.to_numeric(df['PitOutTime'], errors='coerce')
df['PitInTime'] = pd.to_numeric(df['PitInTime'], errors='coerce')

df['PreviousPitInTime'] = df['PitInTime'].shift(1)
df['PreviousStint'] = df['Stint'].shift(1)

df = df[(df['PitOutTime'] > 0) & (df['PreviousPitInTime'] > 0)]

df['PitTimeSeconds'] = df.apply(
    lambda row: row['PitOutTime'] - row['PreviousPitInTime']
    if row['PitOutTime'] > 0 and row['PreviousPitInTime'] > 0 and row['Stint'] != row['PreviousStint'] else None, axis=1
)

df_valid_pit_stops = df[df['PitTimeSeconds'].notnull() & (df['PitTimeSeconds'] > 0)]

pit_time_per_driver = df_valid_pit_stops.groupby(['Driver', 'Team'])['PitTimeSeconds'].sum().reset_index()
pit_time_per_team = df_valid_pit_stops.groupby('Team')['PitTimeSeconds'].sum().reset_index()

pit_time_per_driver['PitTimeMinutes'] = pit_time_per_driver['PitTimeSeconds'] / 60
pit_time_per_team['PitTimeMinutes'] = pit_time_per_team['PitTimeSeconds'] / 60

pit_time_per_driver.to_csv(f'{output_folder}/total_pit_time_per_driver.csv', index=False)
pit_time_per_team.to_csv(f'{output_folder}/total_pit_time_per_team.csv', index=False)

plt.figure(figsize=(14, 6))
sns.barplot(x='Driver', y='PitTimeMinutes', hue='Team', data=pit_time_per_driver, palette="Paired")
plt.title('Total Time Spent in Pits per Driver (Minutes)')
plt.xlabel('Driver')
plt.ylabel('Total Pit Time (minutes)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig(f'{output_folder}/total_pit_time_per_driver_combined.png', dpi=300)
plt.close()

plt.figure(figsize=(12, 6))
sns.barplot(x='Team', y='PitTimeMinutes', data=pit_time_per_team, palette="Set3")
plt.title('Total Time Spent in Pits per Team (Minutes)')
plt.xlabel('Team')
plt.ylabel('Total Pit Time (minutes)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(f'{output_folder}/total_pit_time_per_team_combined.png', dpi=300)
plt.close()

print("Pit stop time recalculated and saved.")
```
