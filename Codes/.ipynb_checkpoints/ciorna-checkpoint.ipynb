```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the lap data
lap_data = pd.read_csv("C:/Users/admin/Desktop/work/lap_2024_cleaned.csv")

# Step 1: Filter the data where only PitInTime is non-zero, indicating a valid pit stop
pit_stops = lap_data[lap_data['PitInTime'] != "0 days 00:00:00"]

# Step 2: Group the data by driver, team, and event (race) to count total pit stops
pit_stop_counts = pit_stops.groupby(['Driver', 'Team', 'EventName']).size().reset_index(name='TotalPitStops')

# Save the pit stop count data to CSV file
pit_stop_counts.to_csv("C:/Users/admin/Desktop/ciorna/pit_stop_counts_only_pitintime.csv", index=False)

# Visualization - Total Pit Stops per Driver
plt.figure(figsize=(10,6))
pit_stop_per_driver = pit_stop_counts.groupby('Driver')['TotalPitStops'].sum()
pit_stop_per_driver.plot(kind='bar')
plt.title('Total Pit Stops per Driver (2024 Season)')
plt.xlabel('Driver')
plt.ylabel('Total Pit Stops')
plt.tight_layout()
plt.savefig("C:/Users/admin/Desktop/ciorna/total_pit_stops_per_driver_only_pitintime.png")
plt.close()

# Visualization - Total Pit Stops per Team
plt.figure(figsize=(10,6))
pit_stop_per_team = pit_stop_counts.groupby('Team')['TotalPitStops'].sum()
pit_stop_per_team.plot(kind='bar', color='orange')
plt.title('Total Pit Stops per Team (2024 Season)')
plt.xlabel('Team')
plt.ylabel('Total Pit Stops')
plt.tight_layout()
plt.savefig("C:/Users/admin/Desktop/ciorna/total_pit_stops_per_team_only_pitintime.png")
plt.close()

print("Pit stop counts and visuals saved in 'C:/Users/admin/Desktop/ciorna' based on PitInTime only.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the converted file
file_path = r'C:\Users\admin\Desktop\work\lap_2024_converted_seconds.csv'
output_folder = r'C:\Users\admin\Desktop\Formula1_output'  # Folder where we'll save files

# Încarcă fișierul CSV convertit
df = pd.read_csv(file_path)

# Convert PitOutTime and PitInTime to numeric
df['PitOutTime'] = pd.to_numeric(df['PitOutTime'], errors='coerce')
df['PitInTime'] = pd.to_numeric(df['PitInTime'], errors='coerce')

# Shift PitInTime to get the previous lap's PitInTime
df['PreviousPitInTime'] = df['PitInTime'].shift(1)
df['PreviousStint'] = df['Stint'].shift(1)

# Filter out rows where PitOutTime or the previous row's PitInTime are 0
df = df[(df['PitOutTime'] > 0) & (df['PreviousPitInTime'] > 0)]

# Calculate pit time only when stint has changed (indicating a valid pit stop)
df['PitTimeSeconds'] = df.apply(
    lambda row: row['PitOutTime'] - row['PreviousPitInTime']
    if row['PitOutTime'] > 0 and row['PreviousPitInTime'] > 0 and row['Stint'] != row['PreviousStint'] else None, axis=1
)

# Filter only valid pit stops
df_valid_pit_stops = df[df['PitTimeSeconds'].notnull() & (df['PitTimeSeconds'] > 0)]

# Group by Driver and Team, sum up the total pit time
pit_time_per_driver = df_valid_pit_stops.groupby(['Driver', 'Team'])['PitTimeSeconds'].sum().reset_index()
pit_time_per_team = df_valid_pit_stops.groupby('Team')['PitTimeSeconds'].sum().reset_index()

# Convert seconds to minutes for better readability
pit_time_per_driver['PitTimeMinutes'] = pit_time_per_driver['PitTimeSeconds'] / 60
pit_time_per_team['PitTimeMinutes'] = pit_time_per_team['PitTimeSeconds'] / 60

# Save the results to CSV files
pit_time_per_driver.to_csv(f'{output_folder}/total_pit_time_per_driver.csv', index=False)
pit_time_per_team.to_csv(f'{output_folder}/total_pit_time_per_team.csv', index=False)

# Plot and save the visualizations
plt.figure(figsize=(14, 6))
sns.barplot(x='Driver', y='PitTimeMinutes', hue='Team', data=pit_time_per_driver, palette="Paired")
plt.title('Total Time Spent in Pits per Driver (Minutes)')
plt.xlabel('Driver')
plt.ylabel('Total Pit Time (minutes)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig(f'{output_folder}/total_pit_time_per_driver_combined.png', dpi=300)
plt.close()

plt.figure(figsize=(12, 6))
sns.barplot(x='Team', y='PitTimeMinutes', data=pit_time_per_team, palette="Set3")
plt.title('Total Time Spent in Pits per Team (Minutes)')
plt.xlabel('Team')
plt.ylabel('Total Pit Time (minutes)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(f'{output_folder}/total_pit_time_per_team_combined.png', dpi=300)
plt.close()

print("Pit stop time recalculated and saved.")
```
